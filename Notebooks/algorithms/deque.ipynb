```c++
#include <iostream>
#include <cstdlib>

using namespace std;
```

```c++
namespace d {
    
    struct node {
        int v;
        struct node *proximo;
        struct node *anterior;
    };

    int _size;
    
    struct node* _frente = NULL;
    struct node* _retaguarda = NULL;

    bool is_fila_empty() {
        return (_frente == NULL) && (_retaguarda == NULL);
    }
    
    int size() {
        return _size;
    }

    void insertFirst(int v) {
        cout << "insertFirst(" << v << ")" << endl;
        if (_frente == NULL) {
            _frente = (struct node *)malloc(sizeof(struct node));
            _frente->anterior = NULL;
            _frente->proximo = NULL;
            _frente->v = v;
            _retaguarda = _frente;
        } 
        else {
            struct node* temp = (struct node *)malloc(sizeof(struct node));
            temp->v = v;
            temp->proximo = _frente;
            _frente->anterior = temp;
            _frente = temp;
        }
        
        _size++;
    }

    void insertLast(int v) {
        cout << "insertLast(" << v << ")" << endl;
        if (_retaguarda == NULL) {
            _retaguarda = (struct node *)malloc(sizeof(struct node));
            _retaguarda->anterior = NULL;
            _retaguarda->proximo = NULL;
            _retaguarda->v = v;
            _frente = _retaguarda;
        } 
        else {
            struct node* temp = (struct node *)malloc(sizeof(struct node));
            temp->v = v;
            temp->anterior = _retaguarda;
            _retaguarda->proximo = temp;
            _retaguarda = temp;
        }
        
        _size++;
    }

    void removeFirst() {
        cout << "removeFirst()" << endl;
        if (_frente == NULL) {
            cout << "Fila vazia" << endl;
            return;
        }
        else {
            struct node* temp = _frente;
            _frente = _frente->proximo;
            if (_frente == NULL) {
                _retaguarda = NULL;
                free(_retaguarda);
            } 
            else {
                _frente->anterior = NULL;
            }
            free(temp);
        }
        
        _size++;
    }

    void removeLast() {
        cout << "removeLast()" << endl;
        if (_retaguarda == NULL) {
            cout << "Fila vazia" << endl;
            return;
        }
        else {
            struct node* temp = _retaguarda;
            _retaguarda = _retaguarda->anterior;
            if (_retaguarda == NULL) {
                _frente = NULL;
                free(_retaguarda);
            } 
            else {
                _retaguarda->proximo = NULL;
            }
            free(temp);
        }
        
        _size++;
    }

    int first() {
        if(is_fila_empty()) {
            cout << "Fila vazia" << endl;
            return -1;
        }
        return _frente->v;
    }

    int last() {
        if(is_fila_empty()) {
            cout << "Fila vazia" << endl;
            return -1;
        }
        return _retaguarda->v;
    }

    void show() {
        struct node* temp = _frente;
        if (is_fila_empty()) {
            cout << "Fila vazia" << endl << endl;
            return;
        }
        int i = 0;
        while (temp != NULL) {
            cout << "(" << i++ << "): " << temp->v << endl;
            temp = temp->proximo;
        }
        cout << endl;
    }

    int run() {
        insertFirst(1);
        show();
        
        insertFirst(2);
        show();
        
        insertFirst(3);
        show();
        
        insertFirst(4);
        show();
        
        insertFirst(5);
        show();
        
        removeFirst();
        show();
        
        removeLast();
        show();
        
        removeFirst();
        show();
        
        removeLast();
        show();
        
        removeLast();
        show();
        
        return 0;
    }

}

int result = d::run();
```
