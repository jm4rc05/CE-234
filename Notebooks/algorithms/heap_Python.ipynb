```python
def min_sift(v, n, i):
    e = 2 * i + 1
    d = 2 * i + 2
    m = i
    if e < n and v[e] < v[i]:
        m = e
    if d < n and v[d] < v[m]:
        m = d
    if m != i:
        v[i], v[m] = v[m], v[i]
        min_sift(v, n, m)
    return

def max_sift(v, n, i):
    e = 2 * i + 1
    d = 2 * i + 2
    m = i
    if e < n and v[e] > v[i]:
        m = e
    if d < n and v[d] > v[m]:
        m = d
    if m != i:
        v[i], v[m] = v[m], v[i]
        max_sift(v, n, m)
    return

def max_heap(v):
    return v[1]

def min_heap(v):
    return v[1]

def extract_max(v):
    m = v[0]
    v[0] = v[len(v) - 1]
    v.pop()
    max_sift(v, len(v), 0)
    return m

def extract_min(v):
    m = v[0]
    v[0] = v[len(v) - 1]
    v.pop()
    min_sift(v, len(v), 0)
    return m

def modify_max(v, k, x):
    v[k] = x
    while(k > 0 and v[k // 2] < v[k]):
        v[k], v[k // 2] = v[k // 2], v[k]
        k //= 2
    max_sift(v, len(v), k)
    return

def modify_min(v, k, x):
    v[k] = x
    while(k > 0 and v[k // 2] > v[k]):
        v[k], v[k // 2] = v[k // 2], v[k]
        k //= 2
    min_sift(v, len(v), k)
    return

def insert_max(v, x):
    v.append(0)
    modify_max(v, len(v) - 1, x)
    return

def insert_min(v, x):
    v.append(0)
    modify_min(v, len(v) - 1, x)
    return
    
def build_min(v):
    for i in range(len(v) // 2 - 1, -1, -1):
        min_sift(v, len(v), i)
    return

def build_max(v):
    for i in range(len(v) // 2 - 1, -1, -1):
        max_sift(v, len(v), i)

def sort_max(v):
    for i in range(len(v) - 1, 0, -1):
        v[i], v[0] = v[0], v[i]
        max_sift(v, i, 0)
    return

def sort_min(v):
    for i in range(len(v) - 1, 0, -1):
        v[i], v[0] = v[0], v[i]
        max_sift(v, i, 0)
    return

def heap_sort_max(v):
    build_max(v)
    for i in range(len(v) - 1, 0, -1):
        v[i], v[0] = v[0], v[i]
        max_sift(v, i, 0)
    return

def heap_sort_min(v):
    build_min(v)
    for i in range(len(v) - 1, 0, -1):
        v[i], v[0] = v[0], v[i]
        min_sift(v, i, 0)
    return

build_max(x := [1, 2, 3, 4, 5, 6, 7, 8, 9])
print(f'heap max: {x}\n')

sort_max(x)
print(f'sorted asc: {x}\n')

build_max(x := [1, 2, 3, 4, 5, 6, 7, 8, 9])
print(f'heap min: {x}\n')

sort_min(x)
print(f'sorted desc: {x}\n')

heap_sort_max(x := [4, 1, 3, 2, 16, 9, 10, 14, 8, 7])
print(f'sorted asc: {x}\n')

heap_sort_min(x := [4, 1, 3, 2, 16, 9, 10, 14, 8, 7])
print(f'sorted desc: {x}')
```
