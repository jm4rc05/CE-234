```c++
#include <iostream>

using namespace std;
```

```c++
namespace fc {
    
    const int LIMITE = 5;

    int _fila[LIMITE] = {0, 0, 0, 0, 0};
    int _frente = 0;
    int _retaguarda = 0;

    int size() {
        int s = (LIMITE - _frente + _retaguarda) % LIMITE;  
        return s;
    }

    bool is_fila_empty() {
        return _retaguarda == _frente;
    }

    void enqueue(int v) {
        cout << "enqueue(" << v << ")" << endl;
        if(size() == LIMITE - 1) {
            cout << "fila cheia " << size() << endl;
            return;
        }
        _fila[_retaguarda] = v;
        _retaguarda = (_retaguarda + 1) % LIMITE;
        cout << "tamanho: " << size() << endl;
    }

    void dequeue() {
        if(is_fila_empty()) {
            cout << "fila vazia" << _retaguarda << "/" << _frente << endl;
            return;
        }
        cout << "dequeue\n";
        _fila[_frente] = 0;
        _frente = (_frente + 1) % LIMITE;
        cout << "tamanho: " << size() << endl;
    }

    int first() {
        if(is_fila_empty()) {
            cout << "fila vazia" << _retaguarda << "/" << _frente << endl;
            return -1;
        }
        return _fila[_frente];
    }

    void show() {
        for(int i = 0; i < LIMITE; i++) {
            cout << "(" << i << "): " << _fila[i] << endl;
        }
        if(is_fila_empty()) {
            cout << "vazia" << endl;
        }    
        cout << "frente    : " << _frente << endl;
        cout << "retaguarda: " << _retaguarda << endl << endl;
    }

    int executa() {
        enqueue(1);
        show();

        enqueue(2);
        show();

        enqueue(3);
        show();

        enqueue(4);
        show();

        enqueue(5);
        show();

        return 0;
    }

}

int result = fc::executa();
```
